# PANDAS DATA STURCTURE

* a) Series - It's a one dimensional array with indexes, it stores a  single column or row
              of data in a Dataframe.

* b) Dataframe - It's tabular spreadsheet like structure representing rows each of which 
                 contains one or multiple columns.


#                              OR simply,


* ->A one-dimensional array(labeled) capable of holding any type of data - Series.
* ->A Two-dimensional array(labeled) structure with columns of potentially different

````python
newdf = pd.DataFrame(np.random.rand(334,5), index=np.arange(334))
````

# FOR THE ABOVE CODE

* Sure! Let's break down the line of code:

```python
newdf = pd.DataFrame(np.random.rand(334, 5), index=np.arange(334))
```

1. `np.random.rand(334, 5)`: 
   - `np.random.rand()` is a NumPy function that generates random numbers.
   - When you pass the arguments `334` and `5` to `np.random.rand()`, it generates an array of shape (334, 5), meaning it creates a 2-dimensional array with 334 rows and 5 columns.
   - Each element in this array is a random number sampled from a uniform distribution over the interval [0, 1).

2. `index=np.arange(334)`:
   - `np.arange(334)` is a NumPy function that generates an array of integers from 0 to 333 (334 numbers in total). 
   - This array is used as the index for the DataFrame `newdf`.
   - So, the DataFrame `newdf` will have 334 rows, and the values in the index will be integers ranging from 0 to 333.

3. `pd.DataFrame(...)`:
   - `pd.DataFrame()` is a pandas function used to create a DataFrame object.
   - It takes the array generated by `np.random.rand(334, 5)` as its data and the array generated by `np.arange(334)` as its index.
   - This results in a DataFrame with 334 rows and 5 columns, filled with random numbers, and indexed from 0 to 333.
