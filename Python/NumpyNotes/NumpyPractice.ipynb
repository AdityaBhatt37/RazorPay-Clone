```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 5-dimensional array
# For simplicity, we'll use small random integers
np.random.seed(0)
arr = np.random.randint(0, 10, size=(2, 2, 2, 2, 4))

# Flatten the array to get a list of points
points = arr.reshape(-1, 4)  # Correct the reshape to match the last dimension

# Since we only have 4 dimensions in the reshaped points, we'll use all four for plotting
# Plot 3D scatter plots for different combinations of the 4 dimensions
fig = plt.figure(figsize=(15, 10))

# Plot for first three dimensions (0, 1, 2)
ax1 = fig.add_subplot(221, projection='3d')
sc1 = ax1.scatter(points[:, 0], points[:, 1], points[:, 2], c=points[:, 3], cmap='viridis')
ax1.set_xlabel('Dim 0')
ax1.set_ylabel('Dim 1')
ax1.set_zlabel('Dim 2')
ax1.set_title('Dimensions 0, 1, 2 (Color: Dim 3)')
fig.colorbar(sc1, ax=ax1, label='Dim 3')

# Plot for dimensions (1, 2, 3)
ax2 = fig.add_subplot(222, projection='3d')
sc2 = ax2.scatter(points[:, 1], points[:, 2], points[:, 3], c=points[:, 0], cmap='viridis')
ax2.set_xlabel('Dim 1')
ax2.set_ylabel('Dim 2')
ax2.set_zlabel('Dim 3')
ax2.set_title('Dimensions 1, 2, 3 (Color: Dim 0)')
fig.colorbar(sc2, ax=ax2, label='Dim 0')

# Plot for dimensions (0, 2, 3)
ax3 = fig.add_subplot(223, projection='3d')
sc3 = ax3.scatter(points[:, 0], points[:, 2], points[:, 3], c=points[:, 1], cmap='viridis')
ax3.set_xlabel('Dim 0')
ax3.set_ylabel('Dim 2')
ax3.set_zlabel('Dim 3')
ax3.set_title('Dimensions 0, 2, 3 (Color: Dim 1)')
fig.colorbar(sc3, ax=ax3, label='Dim 1')

# Plot for dimensions (0, 1, 3)
ax4 = fig.add_subplot(224, projection='3d')
sc4 = ax4.scatter(points[:, 0], points[:, 1], points[:, 3], c=points[:, 2], cmap='viridis')
ax4.set_xlabel('Dim 0')
ax4.set_ylabel('Dim 1')
ax4.set_zlabel('Dim 3')
ax4.set_title('Dimensions 0, 1, 3 (Color: Dim 2)')
fig.colorbar(sc4, ax=ax4, label='Dim 2')

plt.tight_layout()
plt.show()
```
